```python
import requests
zongx2=[]#所有地铁线网址   
from bs4 import BeautifulSoup 
res = requests.get('https://dt.8684.cn/bj')#正则不太会，爬不到所有的数据
html= res.text
soup = BeautifulSoup( html,'html.parser') 
items = soup.find_all(class_='line-a') 
for item in items:
    #print(item['href'])
    zongx2.append('https://dt.8684.cn'+item['href'])
print(zongx2)
```

```python
zz = []#所有线与所属站
subway_connection = {}
for i in range(len(zongx2)):
    zongz[i] = []#每条线所属站
    res = requests.get(zongx2[i])# 返回一个Response对象，赋值给res
    #print(res.encoding)
    html= res.text
    html = html.encode('ISO-8859-1')#编码与解码
    #html = html.decode('utf-8')
    
    soup = BeautifulSoup( html,'html.parser') 
    items = soup.find_all(class_='cl-station') 
    for item in items:
        #print(item.text)
        zongz[i].append(item.text)
    zz.append(zongz[i])
print(zz)
```

```python
subway = []#网络图处理
D={'苹果园':['苹果园', '古城']}
for j in range(len(zz)):
    liwei = [str(i) for i in zongz[j]]
    list1 = zongz[j][-1]
    list2 = zongz[j][0]
    liwei.append(list1)
    litou = [str(i) for i in zongz[j]]
    litou.append(list2)
    subway_connection = {}
    for m in range(len(zongz[j])):
        subway_connection[zongz[j][m]] = []
        subway_connection[zongz[j][m]].append(liwei[m+1])
        subway_connection[zongz[j][m]].append(litou[m-1])
    subway.append(subway_connection)
for time in range(len(subway)-1):
    C={}
    for key1 in list(subway[0]):
        for key2 in list(subway[1]):
            if key1 in list(subway[1]):
                C[key1] = subway[0][key1] + subway[1][key1]
            else:
                C[key1] = subway[0][key1]
                if key2 not in subway[0]:
                    C[key2] = subway[1][key2]
    del subway[0:2]
    subway.append(C)
    subway.reverse()
subway_connection = subway[0]
print(subway_connection)
```

```python
def search(start,destination,connection_graph,sort_candidate):
    pathes = [[start]]
    visitied = set()
    while pathes:#if we find existinng pathes
        path = pathes.pop(0)
        froninter = path[-1]
        if  froninter in visitied: continue
        successors = connection_graph[froninter]
        for city in successors:
            if city in path:continue
            new_path = path + [city]
            pathes.append(new_path)
            if city ==destination: return new_path
        visitied.add(froninter)
        pathes = sort_candidate(pathes)
```
